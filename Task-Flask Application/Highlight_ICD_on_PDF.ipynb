```python
import fitz
import re
from google.colab import files
```

```python
path=r"/content/AI_11_ISC_2_1.pdf"
doc = fitz.open(path)
filename=path.strip(".").split("/")[-1]
filename=filename.strip("/").split(".")[-2]
filename=filename+"_highlight.pdf"
print(filename)
```

```python
icd_pattern=re.compile(r'ICD-\d+-CM:\s*([A-Z]?\d+(?:\.\d+)?(?:\s*,\s*[A-Z]?\d+(?:\.\d+)?)*)')

for page_num in range(len(doc)):
    page = doc[page_num]
    text = page.get_text("text")

    # Find ICD codes in text
    matches = icd_pattern.findall(text)
    if matches:
        for match in matches:
            areas = page.search_for(match)
            for rect in areas:
                highlight = page.add_highlight_annot(rect)
                highlight.update()

# Save the new PDF with highlights
doc.save(filename, garbage=4, deflate=True, clean=True)
doc.close()

files.download(filename)
```

```python
# clcik button wait for user interaction to download file as filename with highlighted file
from google.colab import files
import ipywidgets as widgets
from IPython.display import display
print(filename)
def download_file(button):
    try:
      files.download(filename)
    except Exception as e:
      print("Error: ",e)

button = widgets.Button(description="Download File",button_style='success',tooltip='Click to download file')
button.on_click(download_file)
display(button)
```
