```python
import fitz
import re
from google.colab import files
```

```python
doc = fitz.open("/content/AI_11_ISC_2_1.pdf")
```

```python
icd_pattern=re.compile(r'ICD-\d+-CM:\s*([A-Z]?\d+(?:\.\d+)?(?:\s*,\s*[A-Z]?\d+(?:\.\d+)?)*)')

for page_num in range(len(doc)):
    page = doc[page_num]
    text = page.get_text("text")

    # Find ICD codes in text
    matches = icd_pattern.findall(text)
    if matches:
        for match in matches:
            areas = page.search_for(match)
            for rect in areas:
                highlight = page.add_highlight_annot(rect)
                highlight.update()

# Save the new PDF with highlights
doc.save("highlight_ICD.pdf", garbage=4, deflate=True, clean=True)
doc.close()

files.download("highlight_ICD.pdf")

#button for download when user click button
on_click_download("Highlight_ICD.pdf")
```
