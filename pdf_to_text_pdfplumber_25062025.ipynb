```python
import re
doc="C:/Users/sivaranjani.m/Downloads/AI_11_ISC_2 1.pdf"
def extract_pdf(doc,regex_pattern=None,min_length=50):
    if not doc.lower().endswith(".pdf"):
        return {"status":"error","message":"file is not pdf format"}
    pdf =None
    try:
        cleaned_pages=[]
        pdf=pdfplumber.open(doc)
        cleaned_text=[]
        for page_num, page in enumerate(pdf.pages,start=1):
            text=page.extract_text()
            if regex_pattern:
                cleaned_text=" ".join(re.findall(regex_pattern,text))
            else:
                cleaned_text=re.sub(r"\s"," ",text).strip()
            cleaned_pages.append(cleaned_text)

        return {"status":"success","message":"successfully extracted and stored","data":cleaned_pages}
    except Exception as e:
        return {"status":"error","message":e}
    finally:
        if pdf:
            pdf.close()
        

result=extract_pdf(doc,regex_pattern=r"[A-Za-z0-9@.]+")
if result["status"]=="success":
    for i, page_text in enumerate(result["data"],start=1):
        print("page : ",i)
        print(page_text[:200])
else:
    print("Error:",result['message'])
                
```
